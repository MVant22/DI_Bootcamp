```python
class Phone():
    def __init__(self, phone_number):
        self.phone_number = phone_number
        self.call_history = []
        self.outgoing_messages = []
        self.incoming_messages = []
        
    def call(self, other_phone):
        self.call_history.append(other_phone.phone_number)
        print(f"This number call you {other_phone.phone_number}")
        
    def show_call_history(self):
        if not self.call_history:
            print("Call history is empty")
        else:
            print("Call history")
            for i, number in enumerate(self.call_history, start=1):
                print(f"{i}. {number}")
    
    def send_message(self, other_phone, content):
        message = {
            "to": other_phone.phone_number,
            "from": self.phone_number,
            "content": content
        }
        self.outgoing_messages.append(message)
        other_phone.incoming_messages.append(message)
        print(f"Message sent to {other_phone.phone_number}: {content}")
        
    def show_outgoing_messages(self):
        if not self.outgoing_messages:
            print("No outgoing messages.")
        else:
            print("Outgoing Messages:")
            for msg in self.outgoing_messages:
                print(f"To: {msg['to']} | Content: {msg['content']}")

    def show_incoming_messages(self):
        if not self.incoming_messages:
            print("No incoming messages.")
        else:
            print("Incoming Messages:")
            for msg in self.incoming_messages:
                print(f"From: {msg['from']} | Content: {msg['content']}")

    def show_messages_from(self, sender_number):
        messages = [msg for msg in self.incoming_messages if msg["from"] == sender_number]
        if not messages:
            print(f"No messages from {sender_number}.")
        else:
            print(f"Messages from {sender_number}:")
            for msg in messages:
                print(f"Content: {msg['content']}")

# Create two phones
phone1 = Phone("123-456-7890")
phone2 = Phone("987-654-3210")
phone3 = Phone("555-000-2222")

# Make calls
phone1.call(phone2)
phone2.call(phone1)
phone1.show_call_history()
phone2.show_call_history()

print("\n--- Messaging ---")
# Send messages
phone1.send_message(phone2, "Hello, how are you?")
phone2.send_message(phone1, "I'm good, thanks!")
phone3.send_message(phone1, "Meeting at 10 AM.")

# Show outgoing/incoming
print("\nPhone 1 Outgoing:")
phone1.show_outgoing_messages()

print("\nPhone 1 Incoming:")
phone1.show_incoming_messages()

print("\nPhone 1 Messages From 987-654-3210:")
phone1.show_messages_from("987-654-3210")

print("\nPhone 1 Messages From 555-000-2222:")
phone1.show_messages_from("555-000-2222")

print("\nPhone 1 Messages From 999-999-9999:")
phone1.show_messages_from("999-999-9999")

    
```
